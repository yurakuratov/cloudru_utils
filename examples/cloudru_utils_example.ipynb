## Prepare to run jobs using client_lib

This is the example of sequence of commands that cab be used in bash scripts to run jobs:

1. Init conda and activate the conda environment
2. (Optional) Verify Hugging Face environment variables:
   - Check paths for model and dataset storage
   - Confirm that personal access token is set
3. (Optional) you might check wandb or other env variables here
4. Navigate to the Python script's execution directory
5. Verify current working directory
6. Execute the training command

do not forget to chmod +x ./job_script.sh

```bash
#!/usr/bin/env bash
set -e

eval "$(conda shell.bash hook)"
conda activate /home/jovyan/kuratov/envs/py311_pt2.6_cu12.4

huggingface-cli whoami

cd /home/jovyan/kuratov/data/hidden_capacity/
echo "Current directory: $(pwd)"

# python train.py
# accelerate ... train.py ...
# torchrun ... train.py ...
```

## Running actual jobs

````python
# region SR006
region = 'SR006'
instances = {
    '1gpu': 'a100plus.1gpu.80vG.12C.96G',
    '2gpu': 'a100plus.2gpu.80vG.24C.192G',
    '4gpu': 'a100plus.4gpu.80vG.48C.384G',
    '8gpu': 'a100plus.8gpu.80vG.96C.768G',
}

base_image = 'cr.ai.cloud.ru/aicloud-base-images/py3.11-torch2.4.0:0.0.40'
# conda env is used if job type is pytorch2 (torchrun .. script.py ..)
conda_env = '/home/jovyan/kuratov/envs/py311_pt2.6_cu12.4'
n_gpus = 1
instance_type = instances[f'{n_gpus}gpu']
env_variables = {'TP': f'{n_gpus}',
                 "HOME": '/home/jovyan/kuratov',
                 'HF_HOME': '/home/jovyan/kuratov/data/.cache/huggingface',
                 }
# set job_type to binary for bash scripts or cli commands
job_type = 'binary' #pytorch2  # binary
````
